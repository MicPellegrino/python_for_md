```python
import mdconf_oop as md
import os
```

```python
Lx = 10  # [nm]
Ly = 10  # [nm]
dz = 0.5 # [nm]
```

```python
Lz_list = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
```

```python
walls_conf = dict()
for Lz in Lz_list:
    walls_conf[Lz] = md.Configuration(Lx, Ly, Lz, h="Wall box H="+str(int(Lz)))
    walls_conf[Lz].silica_monolayer(dz, cut_box=True)
    walls_conf[Lz].silica_monolayer(Lz-dz, cut_box=True)
    walls_conf[Lz].output("Vapour/walls_H"+str(Lz).zfill(2)+".gro")
```

```python
N_h2o_list = [1, 10, 100]
scale = 1.00
radius = 0.75
```

```python
for Lz in Lz_list:
    for N in N_h2o_list:
        n_target = int(N*(Lz-1)/4)
        input_file = "Vapour/walls_H"+str(Lz).zfill(2)+".gro"
        output_file = "Vapour/vapour_H"+str(Lz).zfill(2)+"_N"+str(N).zfill(4)+".gro"
        cml = "gmx21_flow insert-molecules"
        cml += " -ci spce_molecule.gro" 
        cml += " -nmol "+str(n_target) 
        cml += " -f "+input_file 
        cml += " -o "+output_file
        cml += " -radius "+str(radius)
        cml += " -scale "+str(scale)
        os.system(cml)
```

```python
os.system("rm \#*#")
dz = 0.75
```

```python
for Lz in Lz_list:
    for N in N_h2o_list:
        n_target = int(N*(Lz-1)/4)
        file_name = "Vapour/vapour_H"+str(Lz).zfill(2)+"_N"+str(N).zfill(4)+".gro"
        conf = md.Configuration()
        conf.input(file_name)
        conf.carve_rectangle([0.0, conf.box_xx], [dz, conf.box_zz-dz])
        conf.output(file_name)
        n = len(conf.res_atomlist["SOL"])/len(conf.res_nomicon["SOL"])
        cml = "gmx21_flow insert-molecules"
        cml += " -ci spce_molecule.gro" 
        cml += " -nmol 1"
        cml += " -f "+file_name 
        cml += " -o "+file_name
        cml += " -radius "+str(radius)
        cml += " -scale "+str(scale)
        while(n<n_target) :
            os.system(cml)
            conf = md.Configuration()
            conf.input(file_name)
            conf.carve_rectangle([0.0, conf.box_xx], [dz, conf.box_zz-dz])
            conf.output(file_name)
            n = len(conf.res_atomlist["SOL"])/len(conf.res_nomicon["SOL"])
```

```python
for Lz in Lz_list:
    for N in N_h2o_list:
        dirname = "H"+str(Lz).zfill(2)+"_N"+str(N).zfill(4)
        os.system("mkdir Vapour/SD/"+dirname)
```
