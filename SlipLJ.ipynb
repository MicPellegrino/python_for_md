```python
import mdconf_oop as md
```

```python
%cd LJSubstrateSlip
%ls
```

```python
box_ref = md.Configuration(10,10,10)
```

```python
sigma_lj = 0.216
r_min = (2**(1/6))*sigma_lj
n_layers = 3
dz = r_min*md.alpha_2
dy = r_min*md.alpha_1
zw0 = 0.5+0.5*3*dy
print("r_min = "+str(r_min))
print("dz = "+str(dz))
print("zw0 = "+str(zw0))
```

```python
box_ref.lj_wall(z_wall=zw0, nk=3, sp=r_min, resname='LJF', cut_box=True)
box_ref.lj_wall(z_wall=zw0+3*dz*(n_layers-2), nk=3, sp=r_min, resname='LJT', cut_box=True)
box_ref.lj_wall(z_wall=zw0+3*dz*(n_layers-1), nk=3, sp=r_min, resname='LJD', cut_box=True)
```

```python
zw0 = (9.5-8*dz)+0.5*9*dy
box_ref.lj_wall(z_wall=zw0, nk=9, sp=r_min, resname='PIS', cut_box=True)
```

```python
box_ref.print_info()
box_ref.output('test.gro')
```
