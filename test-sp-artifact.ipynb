```python
import mdconf_oop as md
import numpy as np
import os
```

```python
%cd /home/michele/python_for_md/TemperatureSptest
```

```python
Ly = 4.7
Lz = 6.0
Lx_vec = [20,40,60,80,100]
```

```python
for Lx in Lx_vec :
    c0 = md.Configuration(Lx,Ly,Lz)
    c0.silica_monolayer(z_wall=Lz-0.5,cut_box=True)
    c0.silica_monolayer(z_wall=0.5,cut_box=True)
    ref_name = 'ref-lx'+str(Lx).zfill(3)+'.gro'
    c0.output(ref_name)
    sol_name = 'sol-lx'+str(Lx).zfill(3)+'.gro'
    cml_solvate = 'gmx solvate -cp '+ref_name+' -cs ../wat_equil.gro -o '+sol_name
    os.system(cml_solvate)
    c1 = md.Configuration()
    c1.input(sol_name)
    c1.carve_rectangle(crop_x=[0.25*Lx,0.75*Lx], crop_z=[0.75,Lz-0.75])
    init_name = sol_name = 'init-lx'+str(Lx).zfill(3)+'.gro'
    c1.output(init_name)
```

```python
for Lx in Lx_vec :
    c2 = md.Configuration()
    init_name = sol_name = 'init-lx'+str(Lx).zfill(3)+'.gro'
    c2.input(init_name)
    print(init_name)
    c2.print_info()
```
